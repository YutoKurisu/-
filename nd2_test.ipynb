```python
with ND2Reader('collagencoat_c1control.nd2') as images:
    intensity_median = np.zeros(images.sizes["v"])
    for i in range(images.sizes["v"]):
        intensity_median[i] = np.median(images[i])
print(intensity_median.mean())
print(np.median(intensity_median))
print(intensity_median)
```

```python
# class
class Student:
    """A simple example class"""
#     sex = "male"
#     ikemendo = 10
    
    def __init__(self, sex, name, ikemendo):
        self.sex = sex
        self.name = name
        self.ikemendo = ikemendo

    def selfintroduce(self):
        return 'I am ' + self.name
```
